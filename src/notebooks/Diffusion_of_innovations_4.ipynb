```python
ooi = "percentage_of_households_owning_ISG_app"
def classify(data):
        data = data[ooi]
        data = data[:,-1]
        classes = np.zeros(data.shape)
        classes[data<20] = 1
        return classes
    
prim_obj = prim.Prim((experiments, outcomes), classify, threshold=0.8)
```
