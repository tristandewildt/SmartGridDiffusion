```python
ooi = "percentage_of_early_majority_with_ISG_appliances"
def classify(data):
    data = data[ooi]
    data = data[:,80]
    classes = np.zeros(data.shape)
    classes[data>30] = 1
    return classes

prim_obj = prim.Prim((experiments, outcomes), classify, threshold=0.8)
```
