```python
desired__nr_lines = 50
nr_cases = experiments.shape[0]
indices = np.arange(0, nr_cases, nr_cases/desired__nr_lines)
```

```python
from analysis.plotting import envelopes, KDE, lines
#from analysis.plotting_util import VIOLIN
results = (experiments, outcomes)

for key, value in outcomes.items():
    if key == 'percentage_of_households_owning_ISG_app':
        fig, axes = lines(results, outcomes_to_show=key, density=KDE,
                      show_envelope=True, experiments_to_show=indices, 
                      titles="")
        plt.show ()
```

```python
import scipy as sp
ooi = "percentage_of_households_owning_ISG_app"
print sp.stats.percentileofscore(outcomes[ooi][:,-1].ravel(), 25)
```

```python
ooi = "percentage_of_households_owning_ISG_app"
def classify(data):
    data = data[ooi]
    data = data[:,-1]
    classes = np.zeros(data.shape)
    classes[data<30] = 1
    return classes

prim_obj = prim.Prim((experiments, outcomes), classify, threshold=0.6)
```

```python
box1 = prim_obj.find_box()
```
